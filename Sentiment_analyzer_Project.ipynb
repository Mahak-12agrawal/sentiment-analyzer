```python
# Simple Sentiment Analyzer (Rule Based)

# Step 1: Create positive and negative word lists
positive_words = ["good", "happy", "love", "excellent", "great", "amazing", "best"]
negative_words = ["bad", "sad", "hate", "worst", "terrible", "awful", "disappointed"]

# Step 2: Take input from user
sentence = input("Enter a sentence: ").lower()

# Step 3: Split sentence into words
words = sentence.split()

# Step 4: Count positive and negative words
positive_count = 0
negative_count = 0

for word in words:
    if word in positive_words:
        positive_count += 1
    elif word in negative_words:
        negative_count += 1

# Step 5: Decide sentiment
if positive_count > negative_count:
    print("Positive Sentiment üòä")
elif negative_count > positive_count:
    print("Negative Sentiment üò°")
else:
    print("Neutral Sentiment üòê")
```
